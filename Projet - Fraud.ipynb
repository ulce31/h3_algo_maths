```python
countries  = []
```

```python
for i in range (len(fraud)):
    ip_address_col = fraud.loc[i, 'ip_address']
    tmp = ip_address[ (ip_address["lower_bound_ip_address"]  <= ip_address_col)  & (ip_address["upper_bound_ip_address"] >= ip_address_col)  ]
    if len(tmp) == 1:
        countries.append(tmp["country"].values[0])
    else :
        countries.append('NA')
fraud['country'] = countries
fraud.head()
```
